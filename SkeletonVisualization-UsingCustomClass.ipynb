# Skeleton Visualization for Cell Type of the Day - 11/11/20

### Initial Imports

```python
import bokeh
import hvplot.pandas
import holoviews as hv
import bokeh.palettes
from bokeh.plotting import figure, show, output_notebook
output_notebook()
from bokeh.io import push_notebook
from SkeletonGraph import SkeletonGraph
from neuprint import Client
from dotenv import load_dotenv
load_dotenv()
import os

authtoken = os.environ.get("api-token")
c = Client('neuprint.janelia.org', dataset='hemibrain:v1.1', token=authtoken)

rois=['SMP(R)','SCL(R)','SLP(R)','CRE(R)']
cellType='SMP573' 

custom_graph = SkeletonGraph(c, cellType, rois)
custom_graph = custom_graph.generateSkeleton()
print(custom_graph)
print("Neuron/Synapse connection count: " + str(custom_graph.neuron_connection_count()))
if(custom_graph.validQuery):
    show(custom_graph.plot)
    push_notebook()
    
cellType='Invalid' 

custom_graph = SkeletonGraph(c, cellType, rois)
custom_graph = custom_graph.generateSkeleton()
print(custom_graph)
print("Neuron/Synapse connection count: " + str(custom_graph.neuron_connection_count()))
if(custom_graph.validQuery):
    show(custom_graph.plot)
    push_notebook()    
    
```
